how many olympics games have been held

```python
%%sql

SELECT COUNT(DISTINCT Games) AS Num_of_games
FROM athletes;
```

list down all olympics games held so far

```python
%%sql

SELECT DISTINCT Games
FROM athletes;
```

mention the total no of nations who partyicipated in each olympic games

```python
%%sql

SELECT Games, COUNT(DISTINCT Team) AS Number_of_Nations
FROM athletes
GROUP BY Games;
```

which year saw the highest and lowest number of countries participating in olympics?

```python
%%sql

SELECT Year, COUNT(Team) AS Total_num_countries
FROM Athletes
GROUP BY Year
ORDER BY total_num_countries DESC
LIMIT 1;
```

```python
%%sql

SELECT Year, MIN(Number_of_nations)
FROM
    (SELECT Year, COUNT(DISTINCT Team))
```

### Number of Males

```python
%%sql

SELECT COUNT(*) FROM athletes WHERE Sex = 'M';
```

### Youngest person

```python
%%sql

SELECT MIN(Age) FROM athletes;
```

### Oldest person

```python
%%sql

SELECT MAX(Age) FROM athletes;
```

### Number of athletes per Sport

```python
%%sql

SELECT Sport, COUNT(*)
FROM athletes
GROUP BY Sport;
```

### Number of athletes per Model

```python
%%sql

SELECT Medal, COUNT(*)
FROM athletes
GROUP BY Medal;
```
